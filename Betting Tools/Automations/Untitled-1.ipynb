```python
import pyautogui
import pytesseract
from PIL import Image
import time
import random
import re

# Set the path to the Tesseract executable if it's not in your system's PATH
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'  # Adjust as necessary

def random_delay():
    """Wait for a random time between 5:15 and 5:45 minutes."""
    time.sleep(random.randint(315, 345))

def capture_and_get_code(region):
    """Capture the screen area and use OCR to get the code."""
    screenshot = pyautogui.screenshot(region=region)
    text = pytesseract.image_to_string(screenshot)
    # Use regular expression to find 10-digit numbers in the OCR result
    match = re.search(r'\b\d{10}\b', text)
    if match:
        return match.group()
    return None

def click_request_code_button(button_position):
    """Click the 'Request Code' button."""
    pyautogui.click(button_position)

def main():
    code_region = (100, 100, 200, 50)  # The region to capture (x, y, width, height), adjust as necessary
    button_position = (150, 150)  # The button's position (x, y), adjust as necessary
    
    while True:
        random_delay()  # Wait for a random interval between 5:15 and 5:45 minutes
        code = capture_and_get_code(code_region)
        if code:
            print(f"Captured code: {code}")
            # You can add your code here to save the code or do something with it
        else:
            print("No code found.")
        
        click_request_code_button(button_position)
        print("Clicked the 'Request Code' button.")

if __name__ == "__main__":
    main()
```
